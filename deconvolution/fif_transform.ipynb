```python
import mne, os
import pandas as pd


destinantion_path = 'C:/Users/mvmigem/Documents/data/project_1/preprocessed/mastoid_ref_csv/'
peak_properties = pd.read_csv(r'C:\Users\mvmigem\Documents\data\project_1\compiled_dataframes\c1_peak_properties.csv')
# Load the 10-20 montage
montage = mne.channels.make_standard_montage('biosemi64')
# Get the list of channel names
electrode_names = montage.ch_names[:]
```

```python

for i in electrode_names:
    spec_path  = os.path.join(destinantion_path, f'raw-{i}')
    if os.path.isdir(spec_path):
        continue

    el_dir  = os.makedirs(spec_path)
    el = i

    for sub in range(1,27)
        if sub == 20: # has no data
            continue
        # Load and read files
        cleaned_data_dir = '/Users/mvmigem/Documents/data/project_1/preprocessed/'
        clean_raw_path = os.path.join(cleaned_data_dir,f'mastoid_raw/main_clean_mastoidref_{sub:02}-raw.fif')
        clean_epo_path = os.path.join(cleaned_data_dir,f'mastoid_ref/main_eventset_mastoidref_{sub:02}-epo.fif')
        raw = mne.io.read_raw_fif(clean_raw_path)
        epochs = mne.read_epochs(clean_epo_path)
        raw.info['bads'] = []
        # Copy event info
        events = epochs.events
        events_id = epochs.event_id
        metadata = epochs.metadata
        subject = metadata['participant'].iloc[1]
        located_quad = metadata['loc_quad'].iloc[1] 
        if located_quad == 0 or located_quad == 2:
            setup = "1"
            setdown = "3"
        else:
            setup = "2"
            setdown = "4"
        # Select the electrode
        raw_el_df =  raw.copy().pick([el]).to_data_frame(long_format = False)
        # selected_el = peak_properties[peak_properties['subject']==subject]['all_pos_peak_channel'].iloc[0]
        # raw_select_df = raw.copy().pick([selected_el]).to_data_frame(long_format = False)
        # Assemble the event df with extra info
        events_df = pd.DataFrame(events[:,0],columns=['sample'])
        events_df['event_codes'] = events[:,2]
        events_df['position'] = events_df['event_codes'] // 10
        events_df['sequence'] = events_df['event_codes'] % 10
        events_df['attention'] = metadata['attention'].reset_index(drop=True)
        events_df['expectation'] = metadata['expected'].reset_index(drop=True)
        events_df['setup'] = setup
        events_df['setdown'] = setdown
        events_df['subject'] = subject
        # events_df['selected_electrode'] = selected_el

        # Save raws
        raw_el_df_path = os.path.join(destinantion_path,f'raw-{el}/raw_mastoidref_{el}_{sub:02}.csv')
        # raw_select_df_path = os.path.join(destinantion_path,f'raw-selected/raw_mastoidref_selected_{sub:02}.csv')
        raw_el_df.to_csv(raw_el_df_path,index = False)
        # raw_select_df.to_csv(raw_select_df_path,index = False)
        # Save events 
        events_csv_path = os.path.join(destinantion_path,f'events/events_{sub:02}.csv')
        events_df.to_csv(events_csv_path,index = False)
```
